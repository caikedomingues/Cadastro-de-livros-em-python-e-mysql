```python
# Import do arquivo que contém as funções que manipulam o banco de dados
import BancoManipulacaoFuncionario

# Acessando o módulo de banco de dados de funcionários.
banco = BancoManipulacaoFuncionario

# Irá controlar as sessões do usuário
id_logado = None
```

Tela de login: Sera necessário ter um login para utilizar qualquer funcionalidade do sistema

```python
# Irá inspecionar o bloco com  o objetivo de capturar possiveis erros
# de execução do trecho
try:
    
    # Ira coletar as entradas dos usuários (id e senha)
    id = input(str("Informe o seu id"))

    senha = input("Informe a sua senha: ")

    # Ira definir se a função de login deve ou não ser chamada.
    status_login = True
    
    if id == "":
    
     # Se o campo de id for vázio, vamos imprimir essa mensagem e
     # atribuir ao status de login o valor False.
     print("O campo de id não pode ser vázio")
     
     status_login = False

    if id.isnumeric() == False:
        
        # Se o valor informado não for um dado numérico, iremos imprmir
        # essa menagem e atribuir ao status de login o valor False que
        # impedirá a chamada da função de login
        print("Por favor não digite letras no campo de id")
        
        status_login = False
    
    if senha == "":
        
        # Se o campo de senha for vázio, vamos imprimir essa mensagem e 
        # e atribuir ao status de login o valor falso
        print("O campo de senha não pode ser vázio")

        status_login = False
        
    if status_login == True:
        
        # Se o status de login for verdadeiro, vamos chamar a função de
        # de funcionários e armazenar o valor retornado na variável
        # id_logado que será validada em todas as funcionalidades que
        # exigirem uma sessão criada no sistema
        id_logado = banco.tela_login_funcionario(id, senha)
        
    
    else:
        
        # Caso contrário, vamos imprimir essa mensagem.
        print("Tente novamente")

except AttributeError as erro:
    
    # Ira tratar erros relacionados a inexistência de funções no módulo
    # do banco de dados
    print("Falha na função de login")
    
```

Tela de cadastro de funcionários: É necessário ter um login para cadastra novos funcionários

```python
# Irá inspecionar o bloco de código com o objetivo de capturar possiveis
# erros.
try:
    
    if id_logado:
        
        # Ira coletar as entradas de usuário (nome e senha). O replace do nome tem como objetivo
        # remover os espaços em branco com o objetivo de conseguirmos utilizar o isalpha para verficar
        # a existência de valores numéricos no campo de nomes.
        nome = input(str("Escreva o nome do funcionário")).replace(' ','')

        senha = input("Crie uma senha para o funcionário (Pode escrever usando quantas letras e números quiser)")

        # Irá definir se a função de cadastro deve ou não ser chamada
        status_cadastro = True

        if nome == "":
            
            # Se o campo de nome for vázio, vamos imprimir essa mensagem
            # e atribuir o valor False ao status de cadastro que irá
            # impedir a chamada da função de cadastro
            print("O campo de nome não pode ser vázio")

            status_cadastro = False
            
        elif nome.isalpha() == False:
            
            # Se o retorno da função isalpha for falso, ou seja, se o valor informado
            # não for do tipo texto, vamos imprimir essa mensagem e atribuir o valor 
            # False ao status de cadastro com o objetivo de impedir a chamada de cadastro
            # de funcionários
            print("O campo de nome não pode possuir números")
            
            status_cadastro = False
        
        if senha == "":
            
            # Se o campo de senha for vázio, vamos imprimir essa mensagem e atribuir
            # o valor False ao status de cadastro com o objetivo de impedir a chamada
            # da função de cadastro.
            print("O campo de senha não pode ser vázio")
            
            status_cadastro = False

        if status_cadastro == True:
            
            # Se o valor do status de cadastro for True, iremos chamar a função
            # de cadastro de funcionários usando os valores de entrada do usuário
            banco.cadastroFuncionario(nome,senha)
            
        else:
            
            # Caso contrário, iremos apenas imprimir essa mensagem
            print("Tente novamente")
    
    else:
        
        print("É necessário ter um login para cadastrar novos funcionários")

except AttributeError as erro:
    
    # Irá tratar erros relacionados a inexistência de funções
    print("Falha na função de cadastro, tente novamente: ", erro)
```

Cadastro de livros

```python
# Irá inspecionar o bloco de código com o objetivo de capturar possiveis erros
# de execução do sistema
try:
    
    # Ira verificar se uma sessão foi criada, ou seja, se o usuário realizou ou não
    # um login no sistema. Basicamente o if irá verificar o retorno da função login
    # que está armazenado na variável id_logado.
    if id_logado:
        
        # Ira coletar as entradas de usuário
        isbn = input("Informe o isbn do livro")

        titulo = input("Titulo do livro")

        # Vamos utilizar o replace para remover os espaços em branco com o objetivo possibilitar
        # a utilização da função isalpha que verifica a existencia de numeros em valores do tipo
        # string (texto)
        autor = input("Autor do livro").replace(" ", "") 

        ano_de_publicacao = input("Ano de Publicação")
 
        quantidade = input("Quantidade")
        
        # Ira verificar se os dados devem ou não ser inseridos no sistema
        status_livro = True


        if isbn == "":
            
            # Se o campo de isbn for vázio, vamos imprimir essa mensagem e
            # atribuiremos ao status do livro o valor false
            print("O campo de isbn não pode ser vázio")
            
            status_livro = False
        
        elif len(isbn) > 6 or len(isbn) < 6:
            
            # Se a quantidade de digitos for maior ou menor que 6, iremos
            # imprimir essa mensagem e atribuiremos ao status do livro,
            # o valor false
            print("O isbn deve conter 6 digitos")
            
            status_livro = False
        
        elif isbn.isnumeric() == False:
            
            # Se o valor do isbn não for numerico iremos imprimir essa mensagem
            # E atribuiremos ao status livro o valor false.
            print("Não digite letras nos campos de isbn")

            status_livro = False
            
        
        if titulo == "":
            
            # Se o campo de titulo for vázio, vamos imprimir essa mensagem
            # e atribuir o valor false ao status do livro
            print("O campo de titulo do livro não pode ser vázio")
            
            status_livro = False
            
        if autor == "":
            
            # Se o campo de autor for vázio, iremos imprimir essa mensagem
            # e atribuiremos o valor false ao status do livro.
            print("O nome do autor não pode ser vázio")
        
        elif autor.isalpha() == False:
            
            # Se o campo de autor tiver a presença de valores numéricos
            # vamos imprimir essa mensagem e artibuir false ao valor do
            # status do livro.
            print("Não digite numeros no nome do autor")
            
            status_livro = False
        
        if ano_de_publicacao.isnumeric() == False:
            
            # Se o campo não for do tipo numérico, iremos imprmir essa mensagem e
            # atribuir ao status do livro o valor False
            print("O ano de publicação naõ deve conter letras")
            
            status_livro = False
        
        elif len(ano_de_publicacao) > 4 or len(ano_de_publicacao) < 4:
            
            # Se a quantidade de digitos for menor ou maior que 4, iremos
            # imprimir essa mensagem e atribuir ao status do livro o valor
            # False
            print("O ano de publicação deve conter apenas 4 digitos")

            status_livro = False
        
        elif ano_de_publicacao == "":
            
            # Se o campo de ano de publicação for vázio, iremos imprimir essa mensagem
            # e atribuir ao status do livro, o valor False
            print("O ano de publicação não pode ser vázio")
            
            status_livro = False
        
        
        
        if quantidade == "":
            
            # Se o campo de quantidade for vázio, vamos imprimir essa mensagem
            # e atribuir ao status do livro o valor false
            print("O campo de quantidade não pode ser vázio")
            
            status_livro = False
        
        elif quantidade.isnumeric() == False:
            
            # Se o campo de quantidade não for númérico, vamos imprimir essa mensagem
            # e atribuir ao status do livro o valor false
            print("Por favor, não digite números")
            
            status_livro = False
        
        
        if status_livro == True:
            
            # Se o status do livro for True, vamos chamar a função de cadastro de livros
            # usando as entradas de usuário como argumento.
            banco.cadastrarLivro(isbn, titulo, autor, ano_de_publicacao, quantidade)
        
        else:
            
            # Se o status do livro for false, vamos apenas imprimir a mensagem e
            # evitar a execução da função
            print("Tente Novamente")
    else:
        
        # Se o usuário tentar cadastrar um livro sem uma sessão criada, iremos
        # imprmir essa mensagem e não possibilitaremos que ele realize essa ação.
        print("Por favor, faça login para cadastrar livros")

except AttributeError as erro:
    
    # Ira lidar com erros de funções inexistentes no módulo de banco de dados
    print("A função não existe no modulo de banco de dados")
```

Tela de atualização de estoques

```python

# Irá inspecionar o bloco de código com o objetivo de capturar possiveis erros de 
# execução.
try:
    
    # Ira verificar se a variável id_logado possui um id retornado na função de login
    # com o objetivo de verificar se o usuário criou uma sessão antes de atualizar o
    # estoque de um livro.
    if id_logado:
        
        # Ira coletar as entradas de usuário (isbn do livro e a quantidade de
        # livros que serão inseridos no estoque).
        isbn = input("Infome o isbn do livro que será atualizado")
        
        quantidade = input("Informe a quantidade de livros que entrarão no estoque")
        
        # Ira verificar se os dados coletados são válidos para envio no banco de dados.
        status_atualizacao = True
        
        if isbn =="":
            
            # se o campo de isbn for vázio, vamos imorimir essa mensagem
            # e vamos atribuir a o status de atualização o valor false
            print("O campo de isbn não pode ser vázio")
            
            status_atualizacao = False
        
        elif len(isbn) > 6 or len(isbn) < 6:
            
            # Se a quantidade de digitos for menor ou maior que 6,
            # iremos imprimir essa mensagem e atribuir ao status de
            # atualização de estoques o valor False
            print("O isbn deve conter apenas 6 digitos")
            
            status_atualizacao = False
        
        elif isbn.isnumeric() == False:
            
            # Se o campo do isbn não for numérico, vamos imprimir essa mensagem
            # e atribuir ao status de atualização o valor False.
            print("Por favor, não digite letras no campo de isbn")
            
            status_atualizacao = False
        
        if quantidade == "":
            
            # Se o campo de quantidade for vázio, vamos imprimir essa mensagem
            # e atribuir ao valor do status de atualização o False
            print("O campo de quantidade não pode ser vázio")
            
            status_atualizacao = False
        
        elif quantidade.isnumeric() == False:
            
            # Se o campo de quantidade não for do tipo numérico, vamos imprimir essa mensagem e atribuir ao status de atualização o valor false
            print("Por favor não digite letras no campo de quantidade")
            
            status_atualizacao = False
        
        if status_atualizacao == True:
            
            # Se o status da atualização for True, vamos chamar a função de atualização
            # de estoques utilizando como argumento o isbn do livro e a quantidade de
            # produtos que serão inseridos no estoque
            banco.atualiza_estoque_livro(isbn, quantidade)
                    
        else:
            
            # Se o status de atualização for false, iremos mostrar essa mensagem
            # e impediremos a chamada do método
            print("Tente Novamente")
            
    else:
        
        # Se o usuário tentar realizar uma atualização de estoque sem
        # criar uma sessão (realizar um login), vamos imprimir essa 
        # mensagem.
        print("É necessário realizar um login para atualizar o estoque")

except AttributeError as erro:
    
    # Ira tratar erros de inexistência de funções no módulo de banco de dados.
    print("Falha na execução da função: ", erro)
```
