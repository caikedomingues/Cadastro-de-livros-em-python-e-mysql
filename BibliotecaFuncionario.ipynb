Tela de Cadastro de funcionários: Caso já tenha um cadastro, pule essa etapa

```python
# Irá inspecionar o bloco de código com o objetivo de capturar possiveis
# erros.
try:
    
    # Ira coletar as entradas de usuário (nome e senha). O replace do nome tem como objetivo
    # remover os espaços em branco com o objetivo de conseguirmos utilizar o isalpha para verficar
    # a existência de valores numéricos no campo de nomes.
    nome = input(str("Escreva o seu nome")).replace(' ','')

    senha = input("Crie uma senha (Pode escrever usando quantas letras e números quiser)")

    # Irá definir se a função de cadastro deve ou não ser chamada
    status_cadastro = True

    if nome == "":
        
        # Se o campo de nome for vázio, vamos imprimir essa mensagem
        # e atribuir o valor False ao status de cadastro que irá
        # impedir a chamada da função de cadastro
        print("O campo de nome não pode ser vázio")

        status_cadastro = False
        
    elif nome.isalpha() == False:
        
        # Se o retorno da função isalpha for falso, ou seja, se o valor informado
        # não for do tipo texto, vamos imprimir essa mensagem e atribuir o valor 
        # False ao status de cadastro com o objetivo de impedir a chamada de cadastro
        # de funcionários
        print("O campo de nome não pode possuir números")
        
        status_cadastro = False
    
    if senha == "":
        
        # Se o campo de senha for vázio, vamos imprimir essa mensagem e atribuir
        # o valor False ao status de cadastro com o objetivo de impedir a chamada
        # da função de cadastro.
        print("O campo de senha não pode ser vázio")
        
        status_cadastro = False

    if status_cadastro == True:
        
        # Se o valor do status de cadastro for True, iremos chamar a função
        # de cadastro de funcionários usando os valores de entrada do usuário
        banco.cadastroFuncionario(nome,senha)
        
    else:
        
        # Caso contrário, iremos apenas imprimir essa mensagem
        print("Tente novamente")

except AttributeError as erro:
    
    # Irá tratar erros relacionados a inexistência de funções
    print("Falha na função de cadastro, tente novamente: ", erro)
```
