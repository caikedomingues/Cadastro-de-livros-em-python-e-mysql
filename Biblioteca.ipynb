Tela de cadastro de clientes (se não tiver um cadastro no sistema, execute esse trecho primeiro para conseguir fazer uma conta)

```python

    
# Inputs que irão coletar dados cadastrais dos clientes como: cpf,
# nome, telefone, email e senha
cpf = input(str("Informe o seu cpf: "))

# Nessa etapa, precisamos substituir os espaços em branco por
# '' com o objetivo de conseguir utilizar o isalpha (que só retorna
# verdadeiro se a string análisada não possuir espaços).
nome = input(str("Informe o seu nome")).replace(' ', '')

telefone = input(str("Informe um telefone para contato:"))

email = input(str("Informe um email: "))

senha = input(str("Crie uma senha (você pode usar letras e numeros e não há quantidade de caracteres)"))

# Variável que irá definir se os dados serão ou não inseridos
# no banco de dados. Ele tera como valor inicial o True (que indica
# que os dados estão prontos para serem inseridos nos dados)  
status_cadastro = True

# Estruturas que irão validar as entradas de usuário no sistema.
        
if cpf == "":
            
    # Se o campo de cpf for vázio, vamos imprimir essa mensagem.
    print("O campo cpf não pode ser vázio")
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False

elif cpf.isnumeric() == False or len(cpf) < 11 or len(cpf) > 11:
            
    # Nessa etapa vamos verificar 2 situações:
    # 1. Se o valor passado pro campo de cpf é menor ou maior que 11 digitos
    # 2. Se o valor passado é realmente um valor numérico. Vamos fazer essa
    # verificação usando o isnumeric retorna TRUE (caso seja numerico) ou False
    # (caso não seja numérico.)
    print("CPF INVÁLIDO: Por favor verifique a quantidade de digitos ou se há a presença de letras")
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False
        
if nome == "":
        
    # Se o campo de nome for vázio, iremos imprimir essa mensagem
    print("O campo de nome não pode ser vázio")
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False
        
elif nome.isalpha() == False:
            
    # O isalpha tem como objetivo verificar se uma string possui ou não
    # números. Ela retorna um valor booleano onde True indica que a string
    # NÃO possui numeros e False indica que a string  POSSUI números.
            
    # Se o isalpha for falso, vamos imprimir essa mensagem. 
    print("Por favor, não digite números nos campos de nome") 
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False
            
if telefone == "":
            
    # Irá verificar se o campo de telefone está vázio. Se essa condição
    # for verdadeira, vamos imprimir essa mensagem.
    print("O campo de telefone não pode ser vázio")
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False
        
elif telefone.isnumeric() == False or len(telefone) > 11 or len(telefone)< 8:
            
    # Nessa etapa iremos validar 2 condições:
    # 1. Se o telefone possui letras nos caracteres informados
    # 2. Se o telefone possui mais de 11 digitos 
    # Se alguma dessas condições for verdadeira, vamos imprimir essa 
    # mensagem.
    print("Telefone inválido, verifique a quantidade de digitos ou se há a presença de letras")
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False
        
        
if email == "":
            
    # Ira verificar se o campo de email está vázio. Se essa condição
    # for verdadeira, vamos imprimir essa mensagem.
    print("O campo de email não pode ser vázio")
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False

        
elif "@" not in email or ".com" not in email:
            
    # Ira verificar se o email não possui os caracteres necessários
    # para o endereço de um email. Se essa condição for verdadeira
    # vamos imprimir essa mensagem. 
    print("Por favor, digite um email válido")
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False
    
if senha == "":
            
    # Ira verificar se o campo de senha está vázio
    print("O campo de senha não pode ser vázio")
    
    # vamos atribuir false ao status com o objetivo de impedir
    # o cadastro do cliente
    status_cadastro = False


# Ira verificar o status do cadastro para definir se os dados devem
# ou não ser inseridos.
if status_cadastro == True:
    
    # Se o status do cadastro for verdadeiro, vamos inserir os dados
    # no banco de dados.
    banco.cadastracliente(cpf, nome, telefone, email, senha)
    
else:
    
    # Se o status for falso, vamos impedir a inserção e imprimir essa mensagem
    print("Tente novamente")


```
