Tela de cadastro de clientes (se não tiver um cadastro no sistema, execute esse trecho primeiro para conseguir fazer uma conta)

```python

        
try:
    
    # Inputs que irão coletar dados cadastrais dos clientes como: cpf,
    # nome, telefone, email e senha
    cpf = input(str("Informe o seu cpf: "))

    # Nessa etapa, precisamos substituir os espaços em branco por
    # '' com o objetivo de conseguir utilizar o isalpha (que só retorna
    # verdadeiro se a string análisada não possuir espaços).
    nome = input(str("Informe o seu nome")).replace(' ', '')

    telefone = input(str("Informe um telefone para contato:"))

    email = input(str("Informe um email: "))

    senha = input(str("Crie uma senha (você pode usar letras e numeros e não há quantidade de caracteres)"))

    # Variável que irá definir se os dados serão ou não inseridos
    # no banco de dados. Ele tera como valor inicial o True (que indica
    # que os dados estão prontos para serem inseridos nos dados)  
    status_cadastro = True

    # Estruturas que irão validar as entradas de usuário no sistema.
            
    if cpf == "":
                
        # Se o campo de cpf for vázio, vamos imprimir essa mensagem.
        print("O campo cpf não pode ser vázio")
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False

    elif cpf.isnumeric() == False or len(cpf) < 11 or len(cpf) > 11:
                
        # Nessa etapa vamos verificar 2 situações:
        # 1. Se o valor passado pro campo de cpf é menor ou maior que 11 digitos
        # 2. Se o valor passado é realmente um valor numérico. Vamos fazer essa
        # verificação usando o isnumeric retorna TRUE (caso seja numerico) ou False
        # (caso não seja numérico.)
        print("CPF INVÁLIDO: Por favor verifique a quantidade de digitos ou se há a presença de letras")
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False
            
    if nome == "":
            
        # Se o campo de nome for vázio, iremos imprimir essa mensagem
        print("O campo de nome não pode ser vázio")
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False
            
    elif nome.isalpha() == False:
                
        # O isalpha tem como objetivo verificar se uma string possui ou não
        # números. Ela retorna um valor booleano onde True indica que a string
        # NÃO possui numeros e False indica que a string  POSSUI números.
                
        # Se o isalpha for falso, vamos imprimir essa mensagem. 
        print("Por favor, não digite números nos campos de nome") 
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False
                
    if telefone == "":
                
        # Irá verificar se o campo de telefone está vázio. Se essa condição
        # for verdadeira, vamos imprimir essa mensagem.
        print("O campo de telefone não pode ser vázio")
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False
            
    elif telefone.isnumeric() == False or len(telefone) > 11 or len(telefone)< 8:
                
        # Nessa etapa iremos validar 2 condições:
        # 1. Se o telefone possui letras nos caracteres informados
        # 2. Se o telefone possui mais de 11 digitos 
        # Se alguma dessas condições for verdadeira, vamos imprimir essa 
        # mensagem.
        print("Telefone inválido, verifique a quantidade de digitos ou se há a presença de letras")
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False
            
            
    if email == "":
                
        # Ira verificar se o campo de email está vázio. Se essa condição
        # for verdadeira, vamos imprimir essa mensagem.
        print("O campo de email não pode ser vázio")
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False

            
    elif "@" not in email or ".com" not in email:
                
        # Ira verificar se o email não possui os caracteres necessários
        # para o endereço de um email. Se essa condição for verdadeira
        # vamos imprimir essa mensagem. 
        print("Por favor, digite um email válido")
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False
        
    if senha == "":
                
        # Ira verificar se o campo de senha está vázio
        print("O campo de senha não pode ser vázio")
        
        # vamos atribuir false ao status com o objetivo de impedir
        # o cadastro do cliente
        status_cadastro = False


    # Ira verificar o status do cadastro para definir se os dados devem
    # ou não ser inseridos.
    if status_cadastro == True:
        
        # Se o status do cadastro for verdadeiro, vamos inserir os dados
        # no banco de dados.
        banco.cadastracliente(cpf, nome, telefone, email, senha)
        
    else:
        
        # Se o status for falso, vamos impedir a inserção e imprimir essa mensagem
        print("Tente novamente")

except AttributeError as erro:
    
    # Irá lidar com erros de funções inexistentes nas classes de banco de dados.
    print("A função não existe na classe BancoManipulacaoCliente: ", erro)


```

Tela de login dos clientes

```python
# Irá inspecionar o bloco de código com o objetivo de capturar possiveis
# erros.
try:
    
    # Irá coletar o cpf e a senha do usuário
    cpf = input("Informe o seu cpf: ")

    senha = input("Informe a sua senha")

    # Variável que irá definir se a execução da função de login
    # deve ou não ser executada.
    status_login = True

    if cpf == "":
        
        # Se o campo de cpf for vázio, vamos mostrar essa mensagem e
        # mudar o status do login para False com o objetivo de evitar
        # a execução da função de login
        print("O campo de cpf não pode ser vázio")
        
        status_login = False

    elif len(cpf) > 11 or len(cpf) < 11:
        
        # Se a quantidade de digitos do cpf for maior ou menor que 11,
        # vamos imprimir essa mensagem e mudar o status do login para
        # False (que evita a execução da função login)
        print("O cpf deve conter 11 digitos")
        
        status_login = False

    elif cpf.isnumeric() == False:
        
        # Se o valor do cpf não for numérico, vamos imprimir essa mensagem
        # e mudar o status do ligin para False com o objetivo de evitar
        # a execução da função de login.
        print("O cpf não deve conter letras")
        
        status_login = False

    if senha == "":
        
        # Se o campo de senha for vázio, vamos imprimir essa mensagem 
        # e mudar o status do login para False com o objetivo de evitar
        # a execução da função de login.
        print("O campo de senha não pode ser vázio")
        
        status_login = False


    if status_login == True:
        
        # Se o status de login for True, iremos executar a função com o cpf
        # e a senha informada.
       cpf_logado =  banco.tela_login_cliente(cpf, senha)

    else:
        
        # Caso contrário, apenas mostraremos essa mensagem  
        print("Tente novamente")

except AttributeError as erro:
    
    # Ira tratar erros de funções inexistentes nas classes de banco de dados
    print("A função chamada não existe na classe BancoManipulaçãoCliente: ", erro)
```

Tela de aluguel: Escolha o isbn do livro que você quer alugar no sistema

```python

# Irá inspecionar o bloco de código com o objetivo de capturar
# possiveis erros.
try:
    
    # Ira verifoicar se uma sessão foi criada, ou seja, se o retorno
    # da função de login está armazenando None (quando não há criação de
    # uma sessão) ou um id (quando há a criação de um id)
    if cpf_logado:
        
        # Ira coletar o isbn do livro que será alugado no sistema
        isbn = input("Informe o isbn do livro que você quer alugar")
        
        # Ira verificar se o isbn informado é válido para a criação do
        # aluguel.
        status_aluguel = True

        if isbn == "":
            
            # Se o campo de isbn for vázio vamos imprimir essa mensagem
            # e atribuir ao status de aluguel o valor False
            print("O campo de isbn não pode ser vázio")
            
            status_aluguel = False
        
        elif isbn.isnumeric() == False:
            
            # Se o campo de isbn não for numérico, vamos imprimir essa mensagem
            # e atribuir ao status de aluguel o valor False
            print("O campo de isbn não pode conter letras")
            
            status_aluguel = False
        
        elif len(isbn) != 6:
            
            # Se o tamanho do isbn for diferente de 6, vamos imprimir essa mensagem
            # e atribuir ao status de aluguel o valor False
            print("O campo deve conter apenas 6 digitos")
            
            status_aluguel = False
        
        if status_aluguel == True:
            
            # Se o status de aluguel for True, ou seja, se o isbn informado
            # for válido, vamos chamar a função de aluguel e utilizar o isbn
            # como argumento da função.
            banco.criarAluguel(cpf_logado, isbn)
        
        else:
            
            # Se o isbn não for válido, iremos imprimir essa mensagem.
            print("Tente Novamente")

    else:
        
        # Se uma sessão não for criada, iremos imprimir essa mensagem
        print("Por favor, realize o login para alugar livros")

except AttributeError as erro: 
    
    # Irá tratar erros relacionados a inexistência de funções no módulo de banco de dados.
    print("Falha na função de criação de aluguéis: ", erro)
            
```

Tela de livros alugados: Verifique os livros que você alugou

```python
# Irá inspecionar o bloco com o objetivo de capturar possiveis erros de execução.
try:
    
    # Ira verificar se o usuário iniciou ou não uma sessão para verificar
    # os seus livros alugados.
    if cpf_logado:
        
        # Se existir uma sessão, vamos chamar a função que mostra os
        # livros que o usuário logado alugou no sistema.
        banco.exibir_livros_alugados(cpf_logado)
    
    else:
        
        # Se uma sessão não for criada, vamos imprimir essa mensagem
        print("Faça login para ver os livros alugados")

except AttributeError as erro:
    
    # Ira tratar erros de inexistência de funções no módulo de banco de dados
    print("Falha na função de consulta: ", erro)
```

Atualização de Dados Cadastrais

```python
# Irá inspecionar o bloco de código com o objetivo de capturar possiveis erros
# de execução do trecho
try:
    
    # Irá verificar se uma sessão foi ou não criada antes de atualizar os dados.
    if cpf_logado:
        
        # Irá coletar as entradas do usuário. Observação:O replace
        # do campo de nome serve para conseguirmos utilizar a função
        # isalpha futuramente.
        nome = input("Atualize o seu nome").replace(" ", "")
                
        telefone = input("Atualize o seu telefone")
        
        email = input("Atualize o seu email")
        
        # Irá verificar se os dados informados estão corretos para
        # a chamada da função.
        status_atualizacao = True
        
        if nome == "":
            
            # Se o campo de nome for vázio, vamos imprimir essa mensagem
            # e atribuir ao status de atualização o valor False    
            print("O campo de nome não pode ser vázio")
            
            status_atualizacao = False

        elif nome.isalpha() == False:
            
            # Se nome do usuário estiver a presença de letras, vamos imprimir essa mensagem
            # e atribuir ao status de atualização o valor False
            print("O campo de nome não pode conter números")
        
        if telefone == "":
            
            # Se o campo de telefone for vázio, vamos imprimir essa mensagem
            # e atribuir ao status de atualização o valor False
            print("O campo de telefone não pode ser vázio")
        
        elif telefone.isnumeric() == False:
            
            # Se o campo de telefone não for do tipo numérico, vamos imprimir essa mensagem
            # e atribuir ao status de atualização o valor False
            print("Por favor não digite letras no campo de nome")
            
            status_atualizacao = False
        
        elif len(telefone) < 11 and len(telefone) < 8:
            
            # Se a quantidade de digitos do telefone for menor que e 8 e 11, 
            # vamos imprimir essa mensageme atribuir ao status de atualização o valor False
            print("O telefone deve conter 11 (celular) ou 8 (telefone fixo) digitos")
            
            status_atualizacao = False
        
        if email == "":
            
            # Se o campo de senha for vázio, vamos imprimir essa mensagem
            # e atribuir ao status de atualização o valor False
            print("O campo de email não pode ser vázio")
            
            status_atualizacao = False
        
        elif "@" not in email or ".com" not in email:
            
            # Se o email informado não conter o @ e o .com, vamos imprimir essa mensagem
            # e atribuir ao status de atualização o valor False
            print("O email deve conter @ e .com em seu endereço")
            
            status_atualizacao = False
        
        if status_atualizacao == True:
            
            # Se o status da atualização for true, significa que os dados
            # estão corretos e podem ser utilizados como argumento da
            # função de atualização de cadastro
            banco.atualizar_dados(nome, telefone, email, cpf_logado)
        
        else:
            
            # Mensagem que será impressa caso os dados não estejam corretos.
            print("Tente Novamente")
            
    else:
        
        # Mensagem que será impressa caso o usuário não crie uma sessão
        print("Crie uma sessão para atualizar dados")

except AttributeError as erro:
    
    # Ira tratar erros de inexistência de funções no módulo de banco de dados.
    print("Falha na atualização dos dados: ", erro)
            
```
